# Checking out Top10 Genres in the library
Creating a pie chart to visualize the top 10 titles by genre
### ⚡ _Drama, Comedy & Action constitute the top 3 genres on the platform_
More interesting plot would be to see year-wise Top5 trend

```python
genre_top10=nflix["genre"].value_counts().nlargest(10)
```

```python
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(genre_top10,labels=genre_top10.index,radius=1.5,autopct='%1.1f%%',textprops={'size':'medium'},wedgeprops={'width':0.50},startangle=90,pctdistance=0.85,labeldistance=0.99)
ax.legend(loc='upper right',shadow=True,bbox_to_anchor=(1.75,1.3))

plt.show()
```

Making the plot even more readable - 
- Using the `explode` param to _split_ the partitions
- Adjusting the percent & labels information for improved readability 

```python
explodelist = []
i = 0
while i<10:
    explodelist.append(0.09)
    i+=1
print(explodelist)
```

```python
fig,ax = plt.subplots(figsize=(5,5))

textprops={'size':'medium'}
wedgeprops={'width':0.40}
autopct='%1.0f%%'
ax.pie(genre_top10,radius=1.5,autopct=autopct,textprops=textprops,wedgeprops=wedgeprops,startangle=90,pctdistance=0.9,explode=explodelist)

ax.legend(labels=genre_top10.index,shadow=True,bbox_to_anchor=(1.75,1.1))

plt.show()
```

```python
genre_top10_d = genre_top10.sort_index().to_dict()
print(genre_top10_d)
type(genre_top10_d)
```

```python
colorcode=[]
for genre,count in genre_top10_d.items():
    if (count <=100):
        colorcode.append('#353652')
    elif (count > 100) and (count <=200):
        colorcode.append('#4c334d')
    elif (count > 200) and (count <=400):
        colorcode.append('#942b3b')
    elif (count > 400) and (count <=500):
        colorcode.append('#643047')
    elif (count > 500) and (count <=600):
        colorcode.append('#942b3b')
    elif (count >600) and (count <=800):
        colorcode.append('#c32530')
    elif count >=1000:
        colorcode.append('#db222a')
print(colorcode)
```

```python
genre_top10_1 = genre_top10.sort_index()
genre_top10_1_labels = genre_top10_1.index.to_list()
type(genre_top10_1_labels)
genre_top10_1_labels
```

```python
fig,ax = plt.subplots(figsize=(15,5))

barchart = ax.bar(genre_top10_1.index,genre_top10_1,color=colorcode,label=genre_top10_1_labels)

ax.set(xlabel="Top 10 genre",ylabel="Count of titles")

ax.legend(labels=genre_top10_1_labels)

ax.bar_label(container=barchart,labels=genre_top10_1)

plt.show()
```
